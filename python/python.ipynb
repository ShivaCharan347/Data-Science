```python
def dtype(df):
    cat = []
    con = []
    for col in df:
        if (df[col].dtypes == "object"):
            cat.append(col)
        else:
            con.append(col)
    return cat, con
cat, con = dtype(movie_df)
table = [cat,con]
print(tabulate({"categorical":cat, "continuous":con}, headers=["Category", "Continuous"]))
```

### TASK 10

```python
t10 = df.sort_values('revenue', ascending= False).head(500)
t10
```

```python
def central_tend(col):
    print(f"Mean of {col}: {round(t10[col].mean(),2)}")
    print(f"Median of {col}: {round(t10[col].median(),2)}")
    print(f"Mode of {col}: {t10[col].mode()[0]}")
```

```python
central_tend('budget')
plt.boxplot('budget', data=t10)
plt.show()
```

```python
central_tend('revenue')
plt.boxplot('revenue', data=t10)
plt.show()
```

```python
central_tend('runtime')
plt.boxplot('runtime', data=t10)
plt.show()
```
